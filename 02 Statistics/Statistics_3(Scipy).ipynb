genel kabul %95 güven aralığında çalışmaktır. bunun dışındaki aralıklarda çalışırken yapmamız gereken z tablosuna bakmaktır.

güven aralığı belirlenir.

%99, z=2.58

sonuç: yaklaşık olarak aralık değeri 232 ile 248 aralığı.
verisini aldığımız interneti kullanan kişilerin %95'i 232 ile 248 saniye arasında internette kalıyor.
verisini aldığımız interneti kullanan kişilerin %99'u 230 ile 250 saniye arasında internette kalıyor.

Satılacak bir ürün için alıcılara ne kadar ücret öderdiniz diye bir soru yöneltiliyor ve verdikleri cevaplar kaydediliyor. 
İstenen; bu ürün için optimum bir fiyat aralığı bulabilmek.

```python
fiyatlar = np.random.randint(10, 110, 1000) #fiyatlar 10 ile 100 arasında 1000 kişiye soruldu.
```

```python
fiyatlar.mean() #otalaması
```

z skoru tablosunu z skoru tablosu yada z skoru olasılık dağılımları yazarak bulabilirsin.

```python
import statsmodels.stats.api as sms
```

```python
#ürüne fiyat tahmin etmek için bu kod kullanılıyor.
sms.DescrStatsW(fiyatlar).tconfint_mean()
```
