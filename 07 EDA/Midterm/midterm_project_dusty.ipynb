```python
def hesapla_gunluk_kalori_alimi():
    # İçeceklerin kalori değerleri (varsayımsal)
    tall_mocha_kalori = float(input("Tall Mocha'nın ortalama kalorisini girin: "))
    medium_coffee_kalori = float(input("Medium Coffee'nin ortalama kalorisini girin: "))

    # Günlük tüketim miktarlarını kullanıcıdan al
    tall_mocha_adet = int(input("Günlük tükettiğiniz Tall Mocha adedini girin: "))
    medium_coffee_adet = int(input("Günlük tükettiğiniz Medium Coffee adedini girin: "))

    # Günlük ortalama kalori alımını hesapla
    gunluk_kalori_alimi = (tall_mocha_kalori * tall_mocha_adet) + (medium_coffee_kalori * medium_coffee_adet)

    print(f"\nGünlük ortalama kalori alımınız: {gunluk_kalori_alimi} kalori")

# Fonksiyonu çağır
hesapla_gunluk_kalori_alimi()
```

```python
# İçeceklerin adları ve kalorilerini görüntüle
for index, row in veri.iterrows():
    icecek_ad = row['Beverage']  # 'Beverage' sütunu içeceğin adını içeriyor, eğer sütun adı farklı ise değiştirmelisiniz
    kalori = row['Calories']  # 'Calories' sütunu içeceğin kalorisini içeriyor, eğer sütun adı farklı ise değiştirmelisiniz
    print(f"{icecek_ad}: {kalori} kalori")
```

```python
# Kullanıcıdan tüketilen içecekleri ve miktarlarını girmesini iste
print("Lütfen tükettiğiniz içecekleri ve miktarlarını girin (Çıkmak için 'q' basın):")
print("Mevcut içecekler:")
print(veri['Beverage'])

secilen_icecekler = []
secilen_miktarlar = []

while True:
    secim = input("Tükettiğiniz bir içecek seçin (Çıkmak için 'q' basın): ")
    if secim.lower() == 'q':
        break

    if secim not in veri['Beverage'].values:
        print("Geçersiz bir içecek seçimi. Lütfen tekrar deneyin.")
        continue

    try:
        secilen_miktar = float(input(f"{secim} miktarını girin (örneğin, 0.5): "))
        secilen_icecekler.append(secim)
        secilen_miktarlar.append(secilen_miktar)
    except ValueError:
        print("Geçersiz bir miktar girdiniz. Lütfen tekrar deneyin.")

# Seçilen içeceklerin kalori değerlerini hesapla
toplam_kalori = 0
for i in range(len(secilen_icecekler)):
    icecek_ad = secilen_icecekler[i]
    miktar = secilen_miktarlar[i]

    kalori = veri.loc[veri['Beverage'] == icecek_ad, 'Calories'].values[0]
    toplam_kalori += (kalori * miktar)

# Günlük toplam kalori alımını ekrana yazdır
print(f"\nGünlük toplam kalori alımınız: {toplam_kalori} kalori")
```
