```python
#icecek boyutlari arasinda en fazla calori farki olan
#icecekleri karsilastirin
df_g = df[df.Beverage_prep == 'Grande'][['Beverage', 'Calories']]
df_t = df[df.Beverage_prep == 'Tall']
df_compare = pd.merge(df_g, df_t, on= 'Beverage', suffixes = ('_grande', '_tall'))
df_compare
df_compare['diff'] = df_compare['Calories_grande'] - df_compare['Calories_tall']
df_compare['diff']
```

```python
#kendi tüketiminize göre günlük ortalama calori aliminizi hesaplayin.
morning = df[df['Beverage']=='Brewed Coffee']['Calories'].mean()
noon = df[df['Beverage']=='Hot Chocolate (Without Whipped Cream)']['Calories'].mean()
evening = df[df['Beverage']=='Mocha']['Calories'].mean()
print('Daily calorie :', round((morning+noon+evening)/3,2))
```

```python
print("Starbucks kahve türlerine göre günlük kahve tüketiminizi giriniz:")

entered_beverages = set()

total_calories = 0

for index, row in df.iterrows():
    beverage_name = row['Beverage']
    
    if beverage_name not in entered_beverages:
        try:
            user_input = float(input(f"Günlük {beverage_name} tüketimini girin (adet): "))
            calories_per_unit = row['Calories']
            total_calories += user_input * calories_per_unit
            entered_beverages.add(beverage_name)
        except ValueError:
            print("Geçersiz giriş. Lütfen sayısal bir değer girin.")

print(f"\nToplam Günlük Kalori Alımı: {total_calories:.2f} kalori")

#Hocam bu soruyu yapamadım boş bırakacaktım, doğru mu bilmiyorum cevabı ChatGPT'den yardım alarak yaptım bilginiz olsun.
```
