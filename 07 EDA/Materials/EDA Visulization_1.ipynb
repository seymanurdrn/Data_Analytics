```python
import numpy as np
x = np.arange(10)
n = len(plt.style.available)
num_rows = int(n/4)

fig = plt.figure(figsize=(15, 15))

for i, s in enumerate(plt.style.available):
    with plt.style.context(s):
        ax = fig.add_subplot(num_rows, 4, i+1)
        for i in range(1, 4):
            ax.plot(x, i * x**2, label='Group %d' % i)
            ax.set_xlabel(s, color='black')
            ax.legend(loc='best')
    
fig.tight_layout()
plt.show()
#matplot'daki farklı stilleri görmüş olduk
```

```python
def salary(x): #adı salary, fonksiyonun değeri x
    if x == 'low':  #x low ise 1 verecek
        return 1
    elif x == 'medium':
        return 2
    elif x == 'high':
        return 3
```

```python
df.salary.apply(lambda x: salary(x))
```

```python
df['salary'] = df.salary.apply(salary)
```

```python
plt.figure(figsize=(14,5))
sns.heatmap(df.corr(numeric_only=True), annot=True )
```
