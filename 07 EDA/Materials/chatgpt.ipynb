```python
import matplotlib.pyplot as plt

plt.hist(df['satisfaction_level'], bins=10, edgecolor='black')
plt.title('Satisfaction Level Distribution')
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(df['satisfaction_level'])
plt.title('Boxplot of Satisfaction Level')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
# Sadece sayısal sütunları seç
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
df_numerical = df[numerical_columns]

# Korelasyon matrisini oluştur
correlation_matrix = df_numerical.corr()

# 'satisfaction_level' ile olan korelasyonu sıralama
correlation_satisfaction = correlation_matrix['satisfaction_level'].sort_values(ascending=False)

# Korelasyon sonuçlarını görüntüleme
print(correlation_satisfaction)
```

```python
import seaborn as sns

sns.scatterplot(x='average_montly_hours', y='satisfaction_level', data=df)
plt.title('Scatter Plot: Average Monthly Hours vs. Satisfaction Level')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['time_spend_company'], df['satisfaction_level'], alpha=0.1)
plt.title('Scatter Plot: Time Spent in Company vs. Satisfaction Level')
plt.xlabel('Time Spent in Company')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
# Koşullara uyan alt grup
selected_group = df[(df['satisfaction_level'] < 0.2) & (df['last_evaluation'] > 0.8)]

# Seçilen alt grup hakkında istatistiksel bilgiler
print(selected_group.describe())

# Seçilen alt grup üzerinde scatter plot
plt.scatter(selected_group['last_evaluation'], selected_group['satisfaction_level'], alpha=0.5)
plt.title('Scatter Plot: Last Evaluation vs. Satisfaction Level (Selected Group)')
plt.xlabel('Last Evaluation')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
selected_group = df[(df['satisfaction_level'] < 0.2) & (df['last_evaluation'] > 0.8)]

# 'salary' ve 'sales' sütunlarına göre gruplama
grouped_data = selected_group.groupby(['salary', 'sales']).size().reset_index(name='count')

# Elde edilen gruplar ve sayıları
grouped_data
```

```python
grouped_data.sort_values(by='count', ascending=False)
```

```python
# satisfaction_level ve last_evaluation sütunları arasındaki farkı hesapla
df['satisfaction_last_evaluation_diff'] = df['satisfaction_level'] - df['last_evaluation']

# Farkı 0.2'den büyük olan gözlemleri seç
selected_rows = df[df['satisfaction_last_evaluation_diff'] > 0.2]

selected_rows
```

```python
# Koşullara uyan gözlemleri seç
selected_rows = df[df['satisfaction_last_evaluation_diff'] > 0.2]

# 'salary' ve 'sales' sütunlarına göre gruplama ve sayıları alma
grouped_data = selected_rows.groupby(['salary', 'sales']).size().reset_index(name='count')

# Sayıları kullanarak sıralama
sorted_grouped_data = grouped_data.sort_values(by='count', ascending=False)

sorted_grouped_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Koşullara uyan gözlemleri seç
selected_rows = df[df['satisfaction_last_evaluation_diff'] > 0.2]

# Kutu grafiği oluştur
plt.figure(figsize=(12, 8))
sns.boxplot(x='sales', y='satisfaction_last_evaluation_diff', hue='salary', data=selected_rows)
plt.title('Boxplot: Satisfaction-Last Evaluation Difference by Salary and Sales')
plt.xlabel('Salary')
plt.ylabel('Satisfaction-Last Evaluation Difference')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 'salary' sütununa göre gruplama
salary_distribution = df['salary'].value_counts()

# Pasta grafiği oluştur
plt.figure(figsize=(8, 8))
plt.pie(salary_distribution, labels=salary_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Distribution of Salary')
plt.show()
```
