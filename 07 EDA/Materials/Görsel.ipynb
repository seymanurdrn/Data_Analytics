```python
import numpy as np
x = np.arange(10)
n = len(plt.style.available)
num_rows = int(n/4)

fig = plt.figure(figsize=(15, 15))

for i, s in enumerate(plt.style.available):
    with plt.style.context(s):
        ax = fig.add_subplot(num_rows, 4, i+1)
        for i in range(1, 4):
            ax.plot(x, i * x**2, label='Group %d' % i)
            ax.set_xlabel(s, color='black')
            ax.legend(loc='best')
    
fig.tight_layout()
plt.show()
```

```python
fig, ax_arr = plt.subplots(nrows=2, ncols=2, sharex=False, sharey=False)

fig.set_figwidth(12)
fig.set_figheight(8)



ax_arr[0, 0].scatter(df.time_spend_company, df.satisfaction_level, marker='+', c='g') 
ax_arr[0, 0].set_title('time_spend_company')


ax_arr[0, 1].scatter(df.average_montly_hours, df.satisfaction_level, marker='+', c='g') 
ax_arr[0, 1].set_title('average_montly_hours')

ax_arr[1, 0].scatter(df.number_project, df.satisfaction_level, marker='+', c='g') 
ax_arr[1, 0].set_title('number_project')

ax_arr[1, 1].scatter(df.salary, df.satisfaction_level, marker='+', c='g') 
ax_arr[1, 1].set_title('salary')

plt.show()
```
